```python
import pandas as pd
df = pd.read_csv("data/u.data")
df.head()
```

```python
df2 = pd.read_csv("data/u.item",encoding="ISO-8859-1")
#df2 = df2[[0,1]]
#df2.columns = ["movieId","title"]
df2.head()
```

```python
df=df.merge(df2, on="movieId")
df.sample(10)
```

```python
def get_top_rated(df, thr=80, no=10):
    df3 = df.groupby("title").rating.mean()
    df4 = pd.DataFrame(df3)
    df4["no"] = df.groupby("title").rating.count()
    df5 = df4[df4["no"]>thr]
    return df5.sort_values(["rating"], ascending=[0]).head(no)
get_top_rated(df, 80, 10)
```

```python
def get_most_reviwed_popu(df, no=10):
    global df6
    df3 = df.groupby("title").rating.sum()
    df6 = pd.DataFrame(df3)
    df6["no"] =df.groupby("title").rating.count()
    maxno = df6["no"].max()
    df6["rating"] = df6["rating"]/maxno
    return df6.sort_values(["rating"], ascending=[0]).head(no)
get_most_reviwed_popu(df)
```

```python
def get_curated(df, thr=80, no=10):
    df3 = df.groupby("title").rating.mean()
    df7 = pd.DataFrame(df3)
    df7["no"] = df.groupby("title").rating.count()
    df8 = df7[df7["no"]>thr]
    maxno = df8["no"].max()
    df8["rating"] += df8["no"]/maxno - 0.5
    return df8.sort_values(["rating"], ascending=[0]).head(no)
get_curated(df, 80, 10)
```

```python
movie_matrix = df.pivot_table(index='userId', columns='title', values='rating')
movie_matrix.sample(5)
```

```python
movie_matrix.shape
```

```python
def get_sim(movie_matrix, movie_name,thr=40, no=10):
    SW_user_rating = movie_matrix[movie_name]
    similar_to_sw=movie_matrix.corrwith(SW_user_rating)
    similar_to_sw.sample(10)
    corr_SW = pd.DataFrame(similar_to_sw, columns=['correlation'])
    corr_SW.dropna(inplace=True)
    dftemp = df.groupby("title").rating.count()
    dftemp = pd.DataFrame(dftemp)
    corr_SW = corr_SW.merge(dftemp, on="title")
    return corr_SW[corr_SW['rating'] > thr].sort_values(by='correlation', ascending=False).head(no)
get_sim(movie_matrix, 'Jumanji (1995)')
```

```python
def get_sim_users_reviews(df, movie_matrix, uid):
    movie_matrix.head()
    SW_user_rating = movie_matrix.loc[uid]
    similar_to_sw=movie_matrix.corrwith(SW_user_rating, axis=1)
    similar_to_sw.head()
    corr_SW = pd.DataFrame(similar_to_sw, columns=['correlation'])
    corr_SW.dropna(inplace=True)
    corr_SW.head()
    
    coUsers = df.groupby("userId").rating.count()
    coUsers = pd.DataFrame(coUsers)
    coUsers.columns = ["no"]
    coUsers.head()
    #dfNew.decribe()
    corr_SW = corr_SW.join(coUsers['no'])
    corr_SW.head()
    corr_SW = corr_SW[corr_SW['no'] > 20].sort_values(by='correlation', ascending=False)
    corr_SW = corr_SW[corr_SW['correlation'] > 0.5].sort_values(by='correlation', ascending=False)
    corr_SW.head(10)
    corr_SW.shape
    corr_SW.head()
    uids = corr_SW.index
    newDf = df[df.userId.isin(uids)]
    newDf.head(10)
    newDf.shape
    return newDf
    #dfNew

def get_sim_user_top_rated(df, movie_matrix, uid):
    newDf = get_sim_users_reviews(df, movie_matrix, uid)
    return get_top_rated(newDf, 10, 10) 
```

```python
def get_year_reviews(df, st_y, end_y):
    df['valid'] = df.title.apply(lambda x: (x.strip()[-5:-1]).isnumeric())
    newDf=df[df['valid']]
    newDf['year'] = newDf.title.apply(lambda x:int((x.strip()[-5:-1])))
    newDf=newDf[newDf['year']>=st_y]
    newDf=newDf[newDf['year']<=end_y]
    return newDf
newdf2 = get_year_reviews(df, 1998, 1998)
##  newdf2[newdf2["year"]==False]
##  newdf2["year"].describe()
  #  print(type(newdf2["year"][0]))
#newdf2.sample(10)
get_top_rated(newdf2, 10,10)
```
